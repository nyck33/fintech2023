```python
'''
1000 butter, 1000 sugar
all other ingredients unlimited
'''
def calculate_num_cookies(butter_required, sugar_required, quantity, sugar=1000, butter=1000):
    num_with_butter = butter // butter_required
    num_with_sugar = sugar // sugar_required
    
    return (min(num_with_butter, num_with_sugar)) * quantity
```

```python
dfcookies['num_makeable'] = dfcookies.apply(lambda row: calculate_num_cookies(row['butter'], row['sugar'], row['quantity']), axis=1)
```
